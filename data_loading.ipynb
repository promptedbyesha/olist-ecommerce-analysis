```python
import pandas as pd
from sqlalchemy import create_engine

engine = create_engine('mysql+pymysql://root:olist1234@127.0.0.1:3306/olist_db')

customers = pd.read_csv('C:/Users/hp/Desktop/project_olist/olist_customers_dataset.csv')
orders = pd.read_csv('C:/Users/hp/Desktop/project_olist/olist_orders_dataset.csv')
order_items = pd.read_csv('C:/Users/hp/Desktop/project_olist/olist_order_items_dataset.csv')
order_payments = pd.read_csv('C:/Users/hp/Desktop/project_olist/olist_order_payments_dataset.csv')

customers.to_sql('customers', engine, if_exists='replace', index=False)
print('customers done:', len(customers), 'rows')

orders.to_sql('orders', engine, if_exists='replace', index=False)
print('orders done:', len(orders), 'rows')

order_items.to_sql('order_items', engine, if_exists='replace', index=False)
print('order_items done:', len(order_items), 'rows')

order_payments.to_sql('order_payments', engine, if_exists='replace', index=False)
print('order_payments done:', len(order_payments), 'rows')

print('ALL DONE')
```

```python
products = pd.read_csv('C:/Users/hp/Desktop/project_olist/olist_products_dataset.csv')
category = pd.read_csv('C:/Users/hp/Desktop/project_olist/product_category_name_translation.csv')

products = products.merge(category, on='product_category_name', how='left')
products.to_sql('products', engine, if_exists='replace', index=False)
print('products done:', len(products), 'rows')
```
